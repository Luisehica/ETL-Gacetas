# Verify file uploaded in s3 bucket

```python
import pandas as pd
from PyPDF2 import PdfFileReader
```

```python
import boto3

s3 = boto3.resource('s3')
bucket = s3.Bucket('gacetas')

files = []
for obj in bucket.objects.all():
    files.append(obj.key)
```

```python
len(files)
```

```python
def extract_number_pages(pdf_path):
    with open(pdf_path, 'rb') as f:
        pdf = PdfFileReader(f, strict=False)
        number_of_pages = pdf.getNumPages()
    return number_of_pages
```

```python
gaceta_url = {}
for obj in bucket.objects.all():
    file_name = obj.key
    print(file_name)
    name = file_name.split(".")[0]
    number = name.split("_")[1]
    gaceta_url[number] = ["http://s3-us-west-1.amazonaws.com/gacetas/"+file_name, extract_number_pages("gacetas/"+file_name)]
```

```python
gaceta_url
```

```python
df_url = pd.DataFrame.from_dict(gaceta_url, orient = 'index')
df_url.tail()
```

```python
df_url.columns
```

```python
df_url = df_url.reset_index()

df_url.rename(columns = {'index':'numero_gaceta', 0: 'url', 1:'pagina'}, inplace = True)

data_types_dict = {'numero_gaceta': int, 'url': str, 'pagina': int}

df_url = df_url.astype(data_types_dict)


df_url.dtypes
```

```python
df_url.index = df_url["numero_gaceta"]
del df_url["numero_gaceta"]
df_url
```

```python
df_url.columns
```

```python
df_gacetas = pd.read_csv("gacetas.csv")
df_gacetas.index = df_gacetas["numero_gaceta"]
del df_gacetas["numero_gaceta"]
df_gacetas.head()
```

```python
df = df_gacetas.join(df_url, on='numero_gaceta', how='right', lsuffix='_left', rsuffix='_right', sort=True)
df
```

```python
df
```

```python
df.to_csv("gacetas_clean.csv")
```

```python
from PyPDF2 import PdfFileReader

def extract_information(pdf_path):
    with open(pdf_path, 'rb') as f:
        pdf = PdfFileReader(f, strict=False)
        information = pdf.getDocumentInfo()
        number_of_pages = pdf.getNumPages()
        pages = []
        for i in range(number_of_pages):
            pages.append(pdf.getPage(i).extractText())
    
    txt = f"""
    Information about: {pdf_path}

    Author: {information.author}
    Creator: {information.creator}
    Producer: {information.producer}
    Subject: {information.subject}
    Title: {information.title}
    Number of pages: {number_of_pages}
    """
    
    print(txt)
    return number_of_pages
```
