```python
import os

list_raw_page = os.listdir("text")
text = []
for raw_page in list_raw_page[0:10]:
    with open("text/"+raw_page) as first_page:
        text.append([raw_text for raw_text in first_page])
    
print(text)
    
```

```python
import spacy
```

```python
# Convert into a single parragraph
extract = text[0][0:100]
newlist = " "
raw_text = newlist.join(extract)
raw_text
```

## Cleaning Text

```python
import re

raw_text = raw_text.lower()
new_text = re.sub("s+", " ", raw_text)
print(new_text)
```

```python
# Carga el modelo pequeño de español
nlp = spacy.load("es_core_news_sm")

# Procesa un texto
doc = nlp(raw_text)

# Itera sobre los tokens
for token in doc:
    # Imprime en pantalla el texto y el part-of-speech tag prediction
    print(token.text, token.pos_, token.dep_, token.head.text)
    # Imprime 
    
```
